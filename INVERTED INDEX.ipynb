```python
import os
import nltk
import nltk.corpus
import pandas as pd
import string
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data= open(p)
        z=data.read()
        
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in z:
            if ele in punc:
                z = z.replace(ele, "")
        
        tokenz=[t for t in z.split()]
        
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        sr= stopwords.words("english")
        clean_tokenz=[]
        for token in tokenz:
            
            wd=token.lower()
            if wd not in sr:
                clean_tokenz.append(wd)
        
        doc_list.append(clean_tokenz)
    return doc_list
    
```
